```python
df.drop('default', axis=1, inplace=True)
df.head()
```

Se elimino la columna Default, por varios motivos, el primero era la cantidad de datos 'Unknown' o no conocidos. El segudo la gran mayoria de los contactos no estaba en default y solamente 3 contactos si lo estaban. De esos 3 contactos, los 3 rechazaron el contratar el deposito. Por lo tanto no nos sirve en nuestro modelo predictivo

```python
print(df['pdays'].value_counts())
```

```python
df['contacted_before'] = df['pdays'].apply(lambda x: 0 if x == 999 else 1)
df.head()
```

En la columna 'contacted_before' hay muchos casos de filas con un valor extremadamente alto en comparacion del resto, a lo que se puede entender que no se ha contactado antes, por lo que creamos una nueva columna, de si se contacto antes o no.

```python
df.drop('pdays', axis=1, inplace=True)
df.head()
```

Se elimino la columna 'pdays' ya que la reemplazamos con 'contacted_before'

```python

df.info()
```

```python
df.drop(columns=['duration'], inplace=True)
df.head()
```

Se elimino la columna duration, puede que haya una tendencia que mientras mas hablan por telefono, esten mas interesados, por lo tanto, es mas posible que contraten el deposito. Para nuestro estudio, tratar de predecir si van a querer contratar el deposito, no podriamos o no deberiamos tener en cuenta el tiempo de llamada, porque es algo que todavia no sabemos o no estamos evaluando, en este caso vamos a buscar mas el perfil de la persona, que cuanto paso hablando por telefono, este interesado o no.

```python
variables_cat = ['job', 'education', 'month', 'poutcome']
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, col in enumerate(variables_cat):
    sns.countplot(data=df, x=col, hue='y', ax=axes[i])
    axes[i].set_title(f'RelaciÃ³n entre {col} e "y"')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Se puede ver que las personas que fueron contactadas antes, tienen mas probabilidad de decir que si, al igual que las que ya conocen el servicio.
En cuanto al mes, el mes de mayo es donde mas se contacto posibles clientes.
En cuanto a job, se puede ver que los admin hay mas posibilidad de que nos acepten el servicio.

```python
df['y'] = pd.factorize(df['y'])[0]
df['job'] = pd.factorize(df['job'])[0]
df['marital'] = pd.factorize(df['marital'])[0]
df['education'] = pd.factorize(df['education'])[0]
df['month'] = pd.factorize(df['month'])[0]
df['poutcome'] = pd.factorize(df['poutcome'])[0]
df['contact'] = pd.factorize(df['contact'])[0]
df['day_of_week'] = pd.factorize(df['day_of_week'])[0]
df['loan'] = pd.factorize(df['loan'])[0]
df['housing'] = pd.factorize(df['housing'])[0]

plt.figure(figsize=(20, 12))
sns.heatmap(df[['y','contacted_before','job','marital','education','month','poutcome','contact','day_of_week',
                'loan','housing','age','campaign','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']].corr(),annot=True, fmt='.2f')
plt.show()
```

Aqui podemos ver y confirmar: contacted_before es el predictor positivo mas fuerte, esto nos quiere decir que hay muchas probabilidades de que las personas que contactamos antes nos digan que si.
nr.employed y euribor3m son los predictores negativos mas fuertes, esto nos indica que la posibilidad de contratar el deposito disminuye.
day_of_week, housing, loan y marital: Todas tienen correlaciones cercanas a 0.00 con la variable objetivo 'y' por lo que es probable que las eliminemos.
euribor3m, nr.employed y emp.var.rate estan muy relacionadas entre si, son variables redundantes por lo que solo nos quedaremos con euribor3m, ya que las otras siguen sus patrones.
campaign tiene una relaciÃ³n casi nula o ligeramente negativa (-0.07) con el Ã©xito (y). Esto indica que llamar demasiadas veces no aumenta la probabilidad de Ã©xito, ademas de la gran cantidad de outliers que puede afectar a nuestro modelo de aprendizaje.

```python
columnas_a_quitar = ['day_of_week', 'housing', 'loan', 'marital', 'nr.employed', 'emp.var.rate']
df_final = df.drop(columns=columnas_a_quitar)
df_final.head()
```

```python
df_final['campaign'] = df_final['campaign'].clip(upper=10)
```

Convertimos todos los datos que son mayores a 10 llamadas, en 10 como maximo. Como vimos antes, de esta manera tendremos menos outliers y no habia relacion en llamar mas veces con probabilidad de exito
